```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

Data1 = {'b' : [0.25, 0.5, 0.75, 1.0, 1.25], 
         'bm25' : [0.24301, 0.25061, 0.25512, 0.25790, 0.24356], 
         'bm25+' : [0.23080, 0.23359, 0.24279, 0.24567, 0.24281], 
         'bm25L' : [0.23281, 0.24236, 0.24682, 0.24806, 0.24594], 
         'atire' : [0.24360, 0.25068, 0.25587, 0.25763, 0.24335]
        }

# BM25
bm25df = pd.DataFrame(Data1, columns=['b', 'bm25'])
plt.plot(bm25df['b'], bm25df['bm25'], color='red', label="BM25", marker='o')

#BM25 Plus
bm25plusdf = pd.DataFrame(Data1, columns=['b', 'bm25+'])
plt.plot(bm25plusdf['b'], bm25plusdf['bm25+'], color='blue', label="BM25+", marker='o')

# BM25L
bm25dfL = pd.DataFrame(Data1, columns=['b', 'bm25L'])
plt.plot(bm25dfL['b'], bm25dfL['bm25L'], color='green', label="BM25L", marker='o')

# BM25 Atire
atiredf = pd.DataFrame(Data1, columns=['b', 'atire'])
plt.plot(atiredf['b'], atiredf['atire'], color='orange', label="Atire", marker='o')


plt.title('Comparison of Ranking Function Performance', fontsize=14)
plt.xlabel('B Parameter', fontsize=14)
plt.ylabel('Mean Average Precision', fontsize=14)
plt.grid(True)


plt.legend()


plt.show()
```
